# HF VLA Playground
Same repo/config setup framework; runs VLA command selection over MuJoCo camera frames.


```python
from __future__ import annotations

from pathlib import Path
import sys

REPO_ROOT = Path.cwd().resolve()
if not (REPO_ROOT / "train_headless.py").exists() and (REPO_ROOT.parent / "train_headless.py").exists():
    REPO_ROOT = REPO_ROOT.parent.resolve()
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

from brains.config import load_runtime_spec
from brains.harnesses import CameraConfig, VLAHarness
from brains.models import HuggingFaceCommandVLA
```

```python
MODEL_ID = "your-org/your-vla-model"
INSTRUCTION = "walk forward"
STEPS = 120
CAMERA_WIDTH = 224
CAMERA_HEIGHT = 224
```

```python
spec = load_runtime_spec(REPO_ROOT / "configs/default.yaml")
camera = CameraConfig(width=CAMERA_WIDTH, height=CAMERA_HEIGHT, fovy_deg=70.0)
harness = VLAHarness(spec=spec, camera=camera)
agent = HuggingFaceCommandVLA(model_id=MODEL_ID, max_new_tokens=24)


def _print_step(frame):
    print(
        f"step {frame['step'] + 1}/{frame['total_steps']} | reward={frame['reward_total']:.4f} | action={frame['vla']['action']}",
        flush=True,
    )


run = harness.run_vla(agent, instruction=INSTRUCTION, steps=STEPS, on_step=_print_step)

{
    "frames": len(run.frames),
    "total_reward": run.total_reward,
    "first_actions": [frame["vla"]["action"] for frame in run.frames[:10]],
}

```
